```python
from qtft_tools import DC_QAOA
from matplotlib import pyplot as plt
```

```python
p1 = DC_QAOA.from_json("p-1.json")
p2 = DC_QAOA.from_json("p-2.json")
p3 = DC_QAOA.from_json("p-3.json")
p4 = DC_QAOA.from_json("p-4.json")
p5 = DC_QAOA.from_json("p-5.json")
p6 = DC_QAOA.from_json("p-6.json")

```

```python
def enery_size(qaoa_groups, p=1):
    approximation_ratios_expected_energy = []
    approximation_ratios_highest_prob_state_energy = []
    for qaoa_group in qaoa_groups:
        if qaoa_group.c_exact == 0:
            continue
        expected_energy = qaoa_group.approximation_ratio()
        highest_prob_state_energy = qaoa_group.cs[0]/qaoa_group.c_exact,
        approximation_ratios_expected_energy.append((expected_energy, qaoa_group.size))
        approximation_ratios_highest_prob_state_energy.append((highest_prob_state_energy, qaoa_group.size))

    energys = []
    sizes = []
    p_list = []
    for energy, size in approximation_ratios_highest_prob_state_energy: 
        #approximation_ratios_expected_energy
        energys.append(energy)
        sizes.append(size)
        p_list.append(p)
    return sizes , p_list, energys

```

```python

data1 = enery_size(p1.groups)[:3:2]
data2 = enery_size(p2.groups)[:3:2]
data3 = enery_size(p3.groups)[:3:2]
data4 = enery_size(p4.groups)[:3:2]
data5 = enery_size(p5.groups)[:3:2]
data6 = enery_size(p6.groups)[:3:2]

plt.scatter(*data1, marker='x', color='r', alpha=0.25)
plt.scatter(*data2, marker='x', color='g', alpha=0.25)
plt.scatter(*data3, marker='x', color='b', alpha=0.25)
plt.scatter(*data4, marker='x', color='r', alpha=0.25)
plt.scatter(*data5, marker='x', color='g', alpha=0.25)
plt.scatter(*data6, marker='x', color='b', alpha=0.25)

```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
for i, p in enumerate([p1, p2, p3, p4, p5, p6]):
    i = i+1
    data1 = enery_size(p.groups)[2]
    plt.scatter(i * np.ones_like(data1), data1, marker='x', color='r', alpha=0.25)
    plt.errorbar(i, np.mean(data1), yerr=np.std(data1), fmt='o', color='k', capsize=5)


# Display the plot
# lt.plot([1,2,3], means, 'kx-')
plt.show()
```
